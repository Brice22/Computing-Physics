```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
x_tolerance = 0.0001
m = 20
a = 0
b = 1
x = []
y = []
def f(x):
    y=x**3+2*x-1
    return y
for i in range(0,m):
    print ('Approximation #', i+1)
    mp = (a+b)/2
    x.append(mp)
    y.append(f(mp))
    print('[', a, ' , ', b, '] contains the solution')
    print(x[i], y[i])
    hwth = (b-a)/2
    if hwth < x_tolerance: 'break'
    if f(a)*f(mp) < 0 : b = mp
    elif f(mp)*f(b) < 0 : a = mp
plt.plot(x, y, 'o')
plt.title('Approximate solution & its y value')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Position", "Velocity", "Acceleration"], loc = "lower right" )
plt.show()
```
